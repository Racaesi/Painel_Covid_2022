Nossa exploração de dados visa avaliar como o COVID-19 afetou diretamente o brasilseiro dando um enfoque maior nas questões de raça/cor, utilizando como base de referência notícias veiculadas, boletins epidemiológicos e dados do SIVEP-Gripe.

## Justificativas 

No início de 2020, os veículos de imprensa noticiaram que o perfil de óbitos pela COVID-19 tinha em sua maioria a cor/raça negra e suas justificativas eram variadas.

Pesquisadores e cientistas provaram incialmente que a cor/raça negra era a mais afetada pelo COVID-19 devido à uma série de fatores sócioeconômicos como menor renda, maior predominância dos negros na periferia, onde a letalidade do vírus é maior, menor acesso à saúde e à vacinação, motivos pelos quais aumenta a prevalência na comunidade negra de comorbidades importantes que juntamente com o COVID elevam a possibilidade de óbito em relação aos outros públicos.

https://www.epsjv.fiocruz.br/podcast/negros-sao-os-que-mais-morrem-por-covid-19-e-os-que-menos-recebem-vacinas-no-brasil

https://www.cnnbrasil.com.br/saude/negros-morrem-40-mais-que-brancos-por-coronavirus-no-brasil/

https://periodicos.ifsul.edu.br/index.php/thema/article/view/1814/1603

###### Essa análise de dados visa responder alguns questionamentos, dentre dos quais, se o perfil de mortalidade pela COVID-19 ainda se mantém o mesmo em relação a raça/cor após 2 anos da análise inicial realizada e veiculada pela imprensa e estudos decorrentes da época.

Cabe ressaltar que no início do monitoramento da catalogação de dados sobre a COVID-19 não existia campos nos Boletins Epidemiológicos que tratavam a questão de raça/cor, mesmo sendo obrigatório desde 2017, pela Portaria nº344 GM/MS, de 01 de        fevereiro de 2017, que dispõe sobre a inclusão e o preenchimento do quesito raça/cor nos formulários dos sistemas de informação de saúde.

Apesar dessa obrigatoriedade da inclusão de cor/raça, essse o quesito não foi elegível para análise de situação epidemiológica da Covid-19 nos primeiros boletins epidemiológicos, ainda que constasse nas fichas de notificação para Síndrome Gripal e para Síndrome Respiratória Aguda Grave (SRAG), ambas utilizadas como instrumentos de registro, monitoramento e avaliação dos casos suspeitos leves e graves da Covid-19, respectivamente, na rede de atenção básica, nos centros de triagem, nas unidades de pronto-atendimento e na rede hospitalar, assim como no formSUs inicialmente elaborado para registros da Covid-19.
###### A incorporação do quesito raça/cor como categoria de análise (Brasil, 2020b) se deu após posicionamentos do GT Racismo e Saúde, da Coalizão Negra e da Sociedade Brasileira de Médicos de Família e Comunidade.

https://www.abrasco.org.br/site/wp-content/uploads/2020/04/Carta-ao-Ministério-da-Saúde-sobre-a-informação-raça_cor-nos-sistemas-de-informação-da-Covid-19-1.pdf

https://www.scielo.br/j/ea/a/LnkzjXxJSJFbY9LFH3WMQHv/?lang=pt

Mesmo com a inclusão de dados à respeito de cor/raça nos Boletins Epidemiológicos subsquentes (*a inserção de raça no Boletim Epidemiológico se deu a partir do BE-09 na 15ª semana Epidemiológica entre 04 e 10 de abril de 2020*), alguns sistemas nacionais de informação em saúde no Brasil refutam as normativas que recomendam a obrigatoriedade da presença e do preenchimento do quesito raça/cor nos instrumentos do Sistema Único de Saúde (SUS), podendo ser problemas relacionados à tecnologia (sistemas legados que dificultam a inserção de novos campos) ou de preenchimento incompleto devido a diversos fatores tais como: A falta de autoidentificação por parte do paciente e o não preenchimento correto por parte do profissional de saúde. 

Num país como o Brasil predonimantemente preto e pardo (segundo o IBGE-2022 o índice de autodeclarados pretos é de 56,1%) é comum ainda em ambientes de saúde uma subnotificação de raça/cor devido à falta de percepção de identidade com a raça/cor negra por parte do paciente; além de uma alta frequência do preenchimento da categoria ‘Ignorado’ no campo raça/cor, que nos leva a acreditar numa desimportância desse dado e expressam o racismo institucional. 

A não inclusão e a incompletude do campo raça/cor dificulta ao enfrentamento do racismo e à promoção da equidade em saúde no país.

**“Na prática, isso quer dizer que, para determinados grupos, as condições de vida afetam de forma a torná-los mais expostos ao adoecimento e à morte. No cenário brasileiro, este grupo com maior risco de adoecer e morrer é representado, considerando a covid, pela população negra”.**

##### Contagem dos registros de raça/cor

Ao contar o quantitativo de registros CS_RACA (1-Branca, 2-Preta, 3-Amarela, 4-Parda, 5-Indígena, 9-Ignorado), verificamos que o quantitativo maior de registros pertence aos autodeclarados brancos, porém o número de dados Ignorados é muito alto, representando cerca de 16.9% dos registros aproximadamente. 

Por isso, seguindo a tratativa dos dados, removemos todos as entradas registradas (linhas) como Ignorado no campo raça  

```python
base.groupby(['CS_RACA']).size()
```

```python
#Porcentagem de registros  por Raça/Cor (1 - Brancos / 2- Pretos / 3- Amarelos / 4- Pardos / 5 - Indígenas / 9 - Ignorados)
def porcentagem_raca(raca):
    cs_porcentagem = base.loc[(base['CS_RACA']==raca),'CS_RACA']
    porcentagem=(cs_porcentagem.count()/len(base.index)) * 100
    return (porcentagem)


# Brancos
print(f'A porcentagem de registros de autodeclarados Brancos é de: {porcentagem_raca(1):.2f}%\n')
# Pretos
print(f'A porcentagem de registros de autodeclarados Pretos é de: {porcentagem_raca(2):.2f}%\n')
# Amarelos
print(f'A porcentagem de registros de autodeclarados Amarelos é de: {porcentagem_raca(3):.2f}%\n')
# Pardos
print(f'A porcentagem de registros de autodeclarados Pardos é de: {porcentagem_raca(4):.2f}%\n')
# Indígenas
print(f'A porcentagem de registros de autodeclarados Indígenas é de: {porcentagem_raca(5):.2f}%\n')
# Ignorados
print(f'A porcentagem de registros Ignorados é de: {porcentagem_raca(9):.2f}%\n')
```

Somando os valores de pretos e pardos obtidos entendemos que o percentual de 39,5%, valor muito diferente dos observados pelo IBGE - PNAD 2022.

```python
pretos_pardos =  porcentagem_raca(2) + porcentagem_raca(4)
print(f'A porcentagem da soma dos registros de Pretos e Pardos é de: {pretos_pardos:.2f}%\n')
```

```python
#Configuração do gráfico comparação IBGE X vacinação raça/cor
grupos = 2
racaibge_percent = (43.0,56.1)
vacina_percent = (indice_vac_brancos,indice_vac_pretos_pardos)

# Criando plot
fig, ax = plt.subplots()
index = np.arange(grupos)
bar_width = 0.25
opacity = 0.8

rects1 = plt.bar(index - bar_width/2, racaibge_percent, bar_width,
alpha=opacity,
color='b',
label='IBGE')

rects2 = plt.bar(index + bar_width/2, vacina_percent, bar_width,
alpha=opacity,
color='r',
label='Vacinados')

plt.xlabel('Raça/Cor')
plt.ylabel('Percentual(%)')
plt.title('Percentual de Raças IBGE x Percentual de Raças Vacinadas (%)')
plt.xticks(index, ('Brancos', 'Pretos/Pardos'))
plt.legend()
ax.bar_label(rects1, fmt = '%.2f')
ax.bar_label(rects2, fmt = '%.2f')
plt.margins(.15,.15)
plt.tight_layout()
plt.show()
```

O gráfico acima nos ajuda a visualizar a discrepância entre os dados relacionados.
